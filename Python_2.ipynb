1. Write a program to accept percentage from the user and display the grade according to the following criteria:

Here's a Python program that accepts a percentage from the user and displays the corresponding grade based on the provided criteria:

```python
# Function to determine grade based on marks
def determine_grade(marks):
    if marks > 90:
        return "A"
    elif marks > 80:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "D"

# Main program
def main():
    # Accept percentage from the user
    try:
        marks = float(input("Enter your percentage: "))
        
        # Ensure the input is within the valid range
        if marks < 0 or marks > 100:
            print("Please enter a valid percentage between 0 and 100.")
        else:
            # Determine and display the grade
            grade = determine_grade(marks)
            print(f"Your grade is: {grade}")
    except ValueError:
        print("Invalid input. Please enter a numeric value.")

# Execute the main function
if __name__ == "__main__":
    main()
```

### Explanation:
1. **determine_grade function**: This function takes the percentage as input and returns the corresponding grade based on the criteria.
2. **main function**:
   - Prompts the user to enter their percentage.
   - Checks if the entered percentage is within the valid range (0 to 100).
   - Calls the `determine_grade` function to get the grade and prints it.
   - Handles invalid inputs by catching `ValueError` and displaying an appropriate message.

### Usage:
- Run the script.
- Enter a percentage when prompted.
- The script will output the corresponding grade based on the given criteria.

2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:

Here's a Python program that accepts the cost price of a bike and displays the road tax to be paid based on the provided criteria:

```python
# Function to determine road tax based on cost price
def determine_road_tax(cost_price):
    if cost_price > 100000:
        return cost_price * 0.15
    elif cost_price > 50000:
        return cost_price * 0.10
    else:
        return cost_price * 0.05

# Main program
def main():
    # Accept cost price from the user
    try:
        cost_price = float(input("Enter the cost price of the bike in Rs: "))
        
        # Ensure the input is a positive value
        if cost_price <= 0:
            print("Please enter a valid positive cost price.")
        else:
            # Determine and display the road tax
            road_tax = determine_road_tax(cost_price)
            print(f"The road tax to be paid is: Rs {road_tax:.2f}")
    except ValueError:
        print("Invalid input. Please enter a numeric value.")

# Execute the main function
if __name__ == "__main__":
    main()
```

### Explanation:
1. **determine_road_tax function**: This function calculates the road tax based on the cost price of the bike:
   - If the cost price is more than 100000, the tax is 15%.
   - If the cost price is more than 50000 but less than or equal to 100000, the tax is 10%.
   - If the cost price is less than or equal to 50000, the tax is 5%.

2. **main function**:
   - Prompts the user to enter the cost price of the bike.
   - Ensures the entered cost price is a positive value.
   - Calls the `determine_road_tax` function to calculate the tax and prints it.
   - Handles invalid inputs by catching `ValueError` and displaying an appropriate message.

### Usage:
- Run the script.
- Enter the cost price of the bike when prompted.
- The script will output the road tax to be paid based on the given criteria.

4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

Here's a Python program that checks how many times a given number can be divided by 3 before it becomes less than or equal to 10:

```python
# Function to count divisions by 3 until number is <= 10
def count_divisions_by_3(number):
    count = 0
    while number > 10:
        number /= 3
        count += 1
    return count

# Main program
def main():
    try:
        # Accept number from the user
        number = float(input("Enter a number: "))
        
        # Ensure the input is positive
        if number <= 0:
            print("Please enter a positive number.")
        else:
            # Calculate and display the number of divisions
            divisions = count_divisions_by_3(number)
            print(f"The number can be divided by 3, {divisions} times before it is less than or equal to 10.")
    except ValueError:
        print("Invalid input. Please enter a numeric value.")

# Execute the main function
if __name__ == "__main__":
    main()
```

### Explanation:
1. **count_divisions_by_3 function**:
   - Initializes a count variable to keep track of the number of divisions.
   - Continuously divides the number by 3 while it is greater than 10.
   - Increments the count for each division.
   - Returns the count once the number is less than or equal to 10.

2. **main function**:
   - Prompts the user to enter a number.
   - Ensures the entered number is positive.
   - Calls the `count_divisions_by_3` function to calculate the number of divisions and prints the result.
   - Handles invalid inputs by catching `ValueError` and displaying an appropriate message.

### Usage:
- Run the script.
- Enter a positive number when prompted.
- The script will output the number of times the given number can be divided by 3 before it becomes less than or equal to 10.

5. Why and When to Use while Loop in Python give a detailed description with example

### Why and When to Use `while` Loop in Python

The `while` loop in Python is a control flow statement that allows code to be executed repeatedly based on a given Boolean condition. The loop will continue to execute the block of code as long as the condition is `True`. Once the condition becomes `False`, the loop stops executing.

#### Reasons to Use `while` Loop:

1. **Indeterminate Iteration**:
   - Use a `while` loop when you need to repeat a block of code but you don't know in advance how many times you need to iterate. This is known as indeterminate iteration, where the loop continues until a certain condition is met.

2. **Condition-Driven Loop**:
   - If you need to keep checking a condition after each iteration and the loop should continue as long as the condition holds true, a `while` loop is appropriate.

3. **Event-Driven Loop**:
   - Use a `while` loop when the loop's continuation is dependent on external events or user input, such as reading data until the end of a file is reached or until a user decides to stop providing input.

### Example of `while` Loop

Here’s an example that demonstrates the use of a `while` loop to simulate a simple banking system where a user can deposit, withdraw, and check balance until they decide to exit.

```python
# Simple banking system using while loop

def display_menu():
    print("\n--- Banking System ---")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

def banking_system():
    balance = 1000  # Initial balance
    while True:
        display_menu()
        choice = input("Enter your choice: ")
        
        if choice == '1':
            print(f"Your current balance is: Rs {balance}")
        elif choice == '2':
            amount = float(input("Enter the amount to deposit: "))
            if amount > 0:
                balance += amount
                print(f"Rs {amount} deposited successfully. New balance: Rs {balance}")
            else:
                print("Invalid deposit amount.")
        elif choice == '3':
            amount = float(input("Enter the amount to withdraw: "))
            if 0 < amount <= balance:
                balance -= amount
                print(f"Rs {amount} withdrawn successfully. New balance: Rs {balance}")
            else:
                print("Invalid withdrawal amount or insufficient balance.")
        elif choice == '4':
            print("Thank you for using the banking system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    banking_system()
```

### Detailed Explanation:

1. **Indeterminate Iteration**:
   - In this example, the number of iterations is not known in advance. The loop continues to prompt the user for actions (check balance, deposit, withdraw, exit) until the user chooses to exit by selecting option `4`.

2. **Condition-Driven Loop**:
   - The condition here is `while True`, which means the loop will run indefinitely until a `break` statement is encountered. The `break` statement is used to exit the loop when the user chooses to exit.

3. **Event-Driven Loop**:
   - The continuation of the loop depends on the user’s input. The loop will keep running as long as the user does not choose to exit.

### Use Cases of `while` Loop:

- **Reading Files Until EOF**:
  ```python
  with open('file.txt', 'r') as file:
      line = file.readline()
      while line:
          print(line, end='')
          line = file.readline()
  ```

- **User Input Until a Condition is Met**:
  ```python
  number = int(input("Enter a positive number (or -1 to stop): "))
  while number != -1:
      if number > 0:
          print(f"You entered: {number}")
      else:
          print("Please enter a positive number.")
      number = int(input("Enter a positive number (or -1 to stop): "))
  ```

- **Waiting for an Event**:
  ```python
  import time

  waiting = True
  start_time = time.time()
  while waiting:
      if time.time() - start_time > 5:  # wait for 5 seconds
          waiting = False
      print("Waiting for event...", end='\r')
      time.sleep(1)
  print("\nEvent occurred!")
  ```

The `while` loop is versatile and can be used in various scenarios where the exact number of iterations is not known beforehand, making it an essential tool for controlling program flow based on dynamic conditions.

6. Use nested while loop to print 3 different pattern.

Here are three different patterns printed using nested `while` loops in Python:

### Pattern 1: Right-Angle Triangle

```
*
**
***
****
*****
```

```python
def pattern1(n):
    i = 1
    while i <= n:
        j = 1
        while j <= i:
            print("*", end="")
            j += 1
        print()  # Move to the next line
        i += 1

# Main function
def main():
    n = 5
    print("Pattern 1: Right-Angle Triangle")
    pattern1(n)

if __name__ == "__main__":
    main()
```

### Pattern 2: Inverted Right-Angle Triangle

```
*****
****
***
**
*
```

```python
def pattern2(n):
    i = n
    while i >= 1:
        j = 1
        while j <= i:
            print("*", end="")
            j += 1
        print()  # Move to the next line
        i -= 1

# Main function
def main():
    n = 5
    print("Pattern 2: Inverted Right-Angle Triangle")
    pattern2(n)

if __name__ == "__main__":
    main()
```

### Pattern 3: Pyramid

```
    *
   ***
  *****
 *******
*********
```

```python
def pattern3(n):
    i = 1
    while i <= n:
        # Print leading spaces
        spaces = n - i
        while spaces > 0:
            print(" ", end="")
            spaces -= 1
        
        # Print stars
        stars = 2 * i - 1
        while stars > 0:
            print("*", end="")
            stars -= 1
        
        print()  # Move to the next line
        i += 1

# Main function
def main():
    n = 5
    print("Pattern 3: Pyramid")
    pattern3(n)

if __name__ == "__main__":
    main()
```

### Explanation:

1. **Pattern 1: Right-Angle Triangle**:
   - The outer `while` loop controls the rows, incrementing from 1 to `n`.
   - The inner `while` loop prints the asterisks (`*`) for each row, incrementing from 1 to the current row number.

2. **Pattern 2: Inverted Right-Angle Triangle**:
   - The outer `while` loop controls the rows, decrementing from `n` to 1.
   - The inner `while` loop prints the asterisks for each row, incrementing from 1 to the current row number.

3. **Pattern 3: Pyramid**:
   - The outer `while` loop controls the rows, incrementing from 1 to `n`.
   - The first inner `while` loop prints leading spaces, decrementing from `n - i` to 0.
   - The second inner `while` loop prints the asterisks, incrementing from 1 to `2 * i - 1`.

### Usage:
- Run the script.
- Each pattern function (`pattern1`, `pattern2`, `pattern3`) is called from the `main` function with `n = 5` to print the patterns.
- Modify `n` to change the size of the patterns.

8. Reverse a while loop to display numbers from 10 to 1

Here's a Python program that uses a `while` loop to display numbers from 10 to 1 in reverse order:

```python
# Function to display numbers from 10 to 1 in reverse order
def reverse_count():
    number = 10
    while number >= 1:
        print(number)
        number -= 1

# Main function
def main():
    reverse_count()

# Execute the main function
if __name__ == "__main__":
    main()
```

### Explanation:
1. **reverse_count function**:
   - Initializes the variable `number` to 10.
   - Uses a `while` loop that continues as long as `number` is greater than or equal to 1.
   - Inside the loop, it prints the current value of `number` and then decrements `number` by 1 with `number -= 1`.

2. **main function**:
   - Calls the `reverse_count()` function to execute the reverse count.

3. **Execution**:
   - The `main()` function is executed when the script is run, triggering the `reverse_count()` function to print the numbers from 10 to 1.

### Usage:
- Run the script to see the numbers from 10 to 1 printed in reverse order.
