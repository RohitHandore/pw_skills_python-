### Q1. Difference in Data Type of Variables `list_` and `array_list`

```python
import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

# Check data types of variables
print("Data type of list_:", type(list_))
print("Data type of array_list:", type(array_list))
```

### Q2. Data Type of Each Element of Variables `list_` and `array_list`

```python
# Data type of each element of list_
print("Data type of each element of list_:")
for item in list_:
    print(type(item))

# Data type of each element of array_list
print("\nData type of each element of array_list:")
for item in array_list:
    print(type(item))
```

### Q3. Data Type of Elements After Changing `array_list` to Integer

```python
# Change dtype of array_list to int
array_list = np.array(object=list_, dtype=int)

# Check data types of elements
print("Data type of each element of list_ (before conversion):")
for item in list_:
    print(type(item))

print("\nData type of each element of array_list (after conversion to int):")
for item in array_list:
    print(type(item))
```

### Q4. Characteristics of Variable `num_array`

```python
import numpy as np

num_list = [[1, 2, 3], [4, 5, 6]]
num_array = np.array(object=num_list)

# (i) Shape
print("Shape of num_array:", num_array.shape)

# (ii) Size
print("Size of num_array:", num_array.size)
```

### Q5. Create Numpy Array of Zeros with Shape (3,3)

```python
import numpy as np

# Create a numpy array of zeros with shape (3,3)
zeros_array = np.zeros((3,3))
print("Zeros Array:")
print(zeros_array)
```

### Q6. Create Identity Matrix of Shape (5,5) Using Numpy

```python
import numpy as np

# Create identity matrix of shape (5,5)
identity_matrix = np.eye(5)
print("Identity Matrix:")
print(identity_matrix)
```
