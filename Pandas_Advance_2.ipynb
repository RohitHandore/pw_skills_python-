### Q1. Print the data present in the second row of the dataframe, df.

```python
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Print data present in the second row
print("Data in the second row:")
print(df.iloc[1])
```

### Q2. Difference between the functions `loc` and `iloc` in `pandas.DataFrame`

- **`loc`**: This function is used for label-based indexing. It selects data by the labels of the rows or columns.
- **`iloc`**: This function is used for integer-based indexing. It selects data by the integer position of the rows or columns.

### Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df, then find the output for both new_df.loc[2] and new_df.iloc[2].

```python
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

print("Output for new_df.loc[2]:")
print(new_df.loc[2])  # loc uses label-based indexing, so it selects the row with label 2

print("\nOutput for new_df.iloc[2]:")
print(new_df.iloc[2])  # iloc uses integer-based indexing, so it selects the third row (index 2)
```

### Q6. Understanding Windows Function in Pandas and Listing Types of Windows Functions

**Windows Function in Pandas**:
- Windows functions in pandas are used for computing a value for each element in a DataFrame column based on a window of preceding and/or following rows.
- These functions perform calculations over a sliding window of data, enabling tasks such as moving averages, rolling sums, and more.

**Types of Windows Functions**:
1. **Rolling Window Functions**: These functions operate on a fixed-size moving window of data.
2. **Expanding Window Functions**: These functions operate on all the data points seen so far.

### Q8. Python program to calculate the difference between two dates in days, hours, and minutes using Pandas timedelta.

```python
import pandas as pd

# Function to calculate difference between two dates
def date_difference():
    date1 = pd.to_datetime(input("Enter the first date (YYYY-MM-DD): "))
    date2 = pd.to_datetime(input("Enter the second date (YYYY-MM-DD): "))

    # Calculate difference using timedelta
    diff = date2 - date1

    print("Difference between the two dates:")
    print("Days:", diff.days)
    print("Hours:", diff.seconds // 3600)
    print("Minutes:", (diff.seconds // 60) % 60)

date_difference()
```
