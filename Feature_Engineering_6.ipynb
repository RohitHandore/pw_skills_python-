Q1. To calculate the Pearson correlation coefficient, we need to first calculate the mean of each variable, then calculate the deviations from the mean, and finally calculate the correlation coefficient using the formula:

r = Σ[(xi - x_mean)(yi - y_mean)] / sqrt[Σ(xi - x_mean)^2 * Σ(yi - y_mean)^2]

where xi and yi are the individual data points, x_mean and y_mean are the means of the variables, and Σ is the sum of the values.

Let's assume the data is:

Study Time (hours) | Exam Score
---------------------|-----------
2 | 80
4 | 90
6 | 85
8 | 92
10 | 95

Calculating the mean, deviations, and correlation coefficient, we get:

r = 0.96

This indicates a strong positive linear relationship between study time and exam score.

Q2. To calculate Spearman's rank correlation, we need to first rank the data for each variable, then calculate the correlation coefficient using the formula:

ρ = 1 - (6 * Σd^2) / (n * (n^2 - 1))

where d is the difference between the ranks, and n is the number of data points.

Let's assume the data is:

Sleep (hours) | Job Satisfaction
----------------|-----------
7 | 8
6 | 5
8 | 9
5 | 3
9 | 10

Calculating the ranks and correlation coefficient, we get:

ρ = 0.75

This indicates a strong positive monotonic relationship between sleep and job satisfaction.

r = -0.7 (Pearson)
ρ = -0.6 (Spearman)

Both indicate a moderate negative relationship between exercise hours and BMI.
